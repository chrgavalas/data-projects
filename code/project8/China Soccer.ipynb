```python
import pandas as pd
```

```python
import matplotlib
%matplotlib inline
```

```python
df = pd.read_excel("China-Investments.xlsx", skiprows=range(0, 5))
```

```python
df.groupby('Sector')['Quantity in Millions'].sum().plot(x='Sector', y='Quantity in Millions', kind='barh')
```

```python
df_entertainment = df[df['Sector'] == 'Entertainment']
```

```python
df_entertainment.groupby('Year')['Quantity in Millions'].sum().plot(kind='bar')
```

```python
df_entertainment
```

```python
df_entertainment.set_value(892, 'Subsector', 'Soccer')
df_entertainment.set_value(993, 'Subsector', 'Soccer')
df_entertainment.set_value(1010, 'Subsector', 'Soccer')
df_entertainment.set_value(1060, 'Subsector', 'Soccer')
df_entertainment.set_value(1161, 'Subsector', 'Soccer')
df_entertainment.set_value(1201, 'Subsector', 'Soccer')
```

```python
df_soccer = df_entertainment[df_entertainment['Subsector'] == 'Soccer']
```

```python
df_soccer
```

```python
df_soccer.sort_values(by='Quantity in Millions').plot(kind='barh', x='Transaction Party', y='Quantity in Millions')
```

```python
df_soccer
```

```python
df_soccer.groupby('Year')['Quantity in Millions'].sum().plot(kind='bar')
```
