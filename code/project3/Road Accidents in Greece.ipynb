```python
import pandas as pd
```

```python
df = pd.read_csv("Greece-accidents.csv")
```

```python
df['LOCATION'].value_counts()
```

```python
df_greece = df[df['LOCATION'] == 'GRC']
```

```python
df_greece2 =  df_greece[df_greece['MEASURE'] == 'NBR']
```

```python
df_greece2.head(48)
```

```python
df_greece3 = df_greece2[df_greece2['SUBJECT'] == 'DEATH']
```

```python
df_greece3.shape
```

```python
import matplotlib
%matplotlib inline
```

```python
df_greece3.sort_values(by='Value', ascending=True).tail(20).plot(kind='barh', y='Value', x='TIME', title='Deaths in road accidents')
```
