```python
import pandas as pd
```

```python
filepath = "NRC-Emotion-Lexicon-v0.92/greek-lex.csv"
emolex_df = pd.read_csv(filepath)
emolex_df
```

```python
emolex_df = emolex_df.rename(columns={'Greek Translation (Google Translate)':'Word'})
```

```python
emolex_df = emolex_df.dropna()
```

```python
emolex_df = emolex_df.drop_duplicates(subset=['Word'])
```

```python
emolex_df
```

```python
import glob
filenames = glob.glob("greek_politics/tsipras*")
filenames
```

```python
speeches = [open(filename).read() for filename in filenames]
len(speeches)
```

```python
speeches_df = pd.DataFrame({
    'text': speeches,
    'filename': filenames
})
speeches_df
```

```python
speeches_df['date'] = speeches_df.filename.str.extract('tsipras:(\d*-\d*)', expand=False)
speeches_df.head(2)
```

```python
speeches_df['datetime'] = pd.to_datetime(speeches_df['date'], format="%Y-%m")
speeches_df.head(2)
```

```python
speeches_df.dtypes
```

```python
emolex_df.Negative.value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
matrix = vec.fit_transform(speeches_df['text'])
results = pd.DataFrame(matrix.toarray(), columns=vec.get_feature_names())
results
```

```python
vec = CountVectorizer(vocabulary=['ανθρωπιστικής', 'κρίση', 'μνημόνια', 
                                  'μνημόνιο', 'εθνική', 'μάχη', 
                                  'αλήθεια', 'εσείς', 'διαπραγμάτευση', 
                                  'ελλάδα', 'μέρκελ'])
matrix = vec.fit_transform(speeches_df['text'])
results = pd.DataFrame(matrix.toarray(), columns=vec.get_feature_names())
results
```

```python
speeches_df['μνημόνιο'] = results['μνημόνιο']
speeches_df['ελλάδα'] = results['ελλάδα']
speeches_df['ανθρωπιστικής'] = results['ανθρωπιστικής']
speeches_df['εσείς'] = results['εσείς']
speeches_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(vocabulary=emolex_df.Word,
                      use_idf=False, norm='l1')
matrix = vec.fit_transform(speeches_df['text'])
vocab = vec.get_feature_names()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df
```

```python
positive_words = emolex_df[emolex_df.Positive == 1]['Word']
speeches_df['positivity'] = wordcount_df[positive_words].sum(axis=1)
negative_words = emolex_df[emolex_df.Negative == 1]['Word']
speeches_df['negativity'] = wordcount_df[negative_words].sum(axis=1)
angry_words = emolex_df[emolex_df.Anger == 1]['Word']
speeches_df['angry'] = wordcount_df[angry_words].sum(axis=1)
speeches_df
```
