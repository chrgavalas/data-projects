```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://www.bbc.co.uk/news')
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
headlines = doc.find_all('h3')

for headline in headlines:
    print(headline.text)
```

```python
links = doc.find_all('a', { 'class': 'gs-c-promo-heading' })

for link in links:
    print(link.text)
```

```python
links = doc.find_all('a', { 'class': 'gs-c-promo-heading' })

for link in links:
    print(link.text)
    print(link['href'])
```

```python
summaries = doc.find_all('p', { 'class': 'gs-c-promo-summary' })

for summary in summaries:
    print(summary.text)
```

```python
stories = doc.find_all('div', { 'class': 'gs-c-promo' })
for story in stories:
    headline = story.find('h3')
    link = story.find('a')
```

```python
stories_list = []
stories = doc.find_all('div', { 'class': 'gs-c-promo' })
for story in stories:
    headline = story.find('h3')
    link = story.find('a')
    summary = story.find('p')
    
    if summary:
        
        story_dict = {
            'headline': headline.text,
            'url': link['href'],
            'summary': summary.text
        }
    else:
        
        story_dict = {
            'headline': headline.text,
            'url': link['href'],
        }    
    
    stories_list.append(story_dict)
```

```python
import pandas as pd
import time
```

```python
stories_df = pd.DataFrame(stories_list)
stories_df.to_csv("bbc-data.csv")
datestring = time.strftime("%Y-%m-%d-%H-%M")
filename = "bbc-data-" + datestring + ".csv"
stories_df.to_csv(filename, index=False)
```
