```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

%matplotlib inline
```

```python
df = pd.read_csv("pm25_annual_fixed.csv")
df.head()
```

```python
df.columns
```

```python
df['99th Percentile']
```

```python
points = df.apply(lambda row: Point(row.Longitude, row.Latitude), axis=1)
points
```

```python
air_stations = gpd.GeoDataFrame(df, geometry=points)
air_stations.crs = {'init': 'epsg:4326'}
air_stations.plot()
```

```python
states = gpd.read_file("cb_2016_us_state_500k/cb_2016_us_state_500k.shp")
states.head()
```

```python
continental = states[states['NAME'] != 'Alaska']
```

```python
continental = continental[continental['NAME'] != 'Hawaii']
```

```python
continental = continental[continental['NAME'] != 'Puerto Rico']
```

```python
continental = continental[continental['NAME'] != 'American Samoa']
```

```python
continental = continental[continental['NAME'] != 'Guam']
```

```python
continental = continental[continental['NAME'] != 'Commonwealth of the Northern Mariana Islands']
```

```python
continental = continental[continental['NAME'] != 'United States Virgin Islands']
```

```python
continental = continental.to_crs(air_stations.crs)
```

```python
ax = continental.plot(figsize=(25,10), color='lightgrey', linewidth=0.25, edgecolor='white')
air_stations.plot(column='99th Percentile', ax=ax, markeredgewidth=5, cmap='GnBu', markersize=5)
ax.axis('off')
```
